```python
# Import libraries
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
import seaborn as sns

# Inline indicates graphs should be presented as a cell output
%matplotlib inline
```

```python
# Load data
cars_data = pd.read_csv('Cars93.csv', index_col = 0, keep_default_na=False)
# Note the additional keep_default_na parameter which is set to False. This ensures
# that pandas does not interpret the None string in the AirBags column as NaN.
# Display the first few rows of the dataset to understand its structure
#cars_data.head(), cars_data.info()
```

1. Boxplot

```python
# 1. Box plot for Rev.per.mile by manufacturer for Audi, Hyundai, Suzuki, and Toyota

# Setting a consistent style for the plots
sns.set(style="whitegrid")

manufacturers = ["Audi", "Hyundai", "Suzuki", "Toyota"]
subset = cars_data[cars_data['Manufacturer'].isin(manufacturers)]

plt.figure(figsize=(10, 6))
sns.boxplot(data=subset, x="Manufacturer", y="Rev.per.mile")
plt.title("Box Plot of Revs per Mile by Manufacturer")
plt.ylabel("Revs per Mile")
plt.xlabel("Manufacturer")
plt.show()

# Extracting the manufacturer with the car that has the highest Rev.per.mile
highest_revs_car = subset.loc[subset['Rev.per.mile'].idxmax(), ['Manufacturer', 'Model', 'Rev.per.mile']]
highest_revs_car
```

Boxplot question:
Which of these manufacturers has the car with the highest revs per mile?

Boxplot answer: Toyota Tercel has the highest revs per mile, at 3,505.

2. Histogram ¶

```python
# Histogram of MPG
# 2. Histogram of MPG in the city and highway on the same axis
plt.figure(figsize=(10, 6))
sns.histplot(cars_data['MPG.city'], color='blue', label='City MPG', kde=False, bins=15, alpha=0.7)
sns.histplot(cars_data['MPG.highway'], color='green', label='Highway MPG', kde=False, bins=15, alpha=0.7)
plt.title("Histogram of MPG: City vs Highway")
plt.xlabel("MPG")
plt.ylabel("Frequency")
plt.legend()
plt.show()

# Comparing fuel efficiency in city vs highway
mean_mpg_city = cars_data['MPG.city'].mean()
mean_mpg_highway = cars_data['MPG.highway'].mean()

mean_mpg_city, mean_mpg_highway
```

Histogram question:
Is it generally more fuel efficient to drive in the city or on the highway?

Histogram answer: The histogram illustrates that highway MPG tends to be higher than city MPG
This confirms that cars are generally more fuel-efficient when driving on the highway compared to the city.


3. Lineplot ¶

```python
# 3. Line plot for Wheelbase vs Turning Circle
plt.figure(figsize=(10, 6))
sns.lineplot(data=cars_data, x='Wheelbase', y='Turn.circle', marker='o', linestyle='-', color='purple')
plt.title("Relationship Between Wheelbase and Turning Circle")
plt.xlabel("Wheelbase (inches)")
plt.ylabel("Turning Circle (feet)")
plt.show()

# Observing the correlation between Wheelbase and Turning Circle
wheelbase_turn_circle_corr = cars_data[['Wheelbase', 'Turn.circle']].corr().iloc[0, 1]
wheelbase_turn_circle_corr
```

Lineplot question:
What is this relationship? What happens when the wheelbase gets larger?

Lineplot answer: The correlation coefficient, 0.7233, indicates a strong positive linear relationship between the Wheelbase and the Turning Circle. This means that as the wheelbase increases, the turning circle also tends to increase.

4. Barplot ¶

```python
# Barplot for mean horsepower
plt.figure(figsize=(12, 6))
mean_horsepower_by_type = cars_data.groupby('Type')['Horsepower'].mean().sort_values()

sns.barplot(x=mean_horsepower_by_type.index, 
            y=mean_horsepower_by_type.values, 
            hue=mean_horsepower_by_type.index,  # Assign `x` to `hue`
            dodge=False,  # Ensures bars don't separate for each hue
            palette="viridis")  # Keep the palette
plt.title("Mean Horsepower by Car Type")
plt.xlabel("Car Type")
plt.ylabel("Mean Horsepower")
plt.xticks(rotation=45)
plt.show()
```

## Barplot question:
Does a larger car mean more horsepower?

Barplot answer: The Barplot illustrates that the larger the car has more horsepower.
